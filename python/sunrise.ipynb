```python
import math
import numpy as np

import matplotlib.pyplot as plt
```

```python
n=np.arange(1,366).repeat(8,0)
```

```python
Lat=40
n=np.arange(1,366).repeat(8,0)
delta=23.45*np.sin((n-80)*3.14/180*360/365)*3.14/180
omegarise=-np.arccos(-np.tan(Lat*np.pi/180)*np.tan(delta))
omegaset=np.arccos(-np.tan(Lat*np.pi/180)*np.tan(delta))
sunrise=12+omegarise*180/np.pi/15
sunset=12+omegaset*180/np.pi/15

# print(delta)
print(sunrise)
print(sunset)
# np.savetxt('D:\Desktop\sunrise.txt', sunrise, delimiter=',', fmt='%.2f')
# np.savetxt('D:\Desktop\sunset.txt', sunset, delimiter=',', fmt='%.2f')
```

```python
plt.plot(n,sunrise)
plt.plot(n,sunset)
```

```python
localtime=9
DaySize=3
if((localtime < sunrise) or(sunset < localtime)):
    CorrectionFactor=0
elif((abs(localtime-sunrise) < DaySize)):
    CorrectionFactor=abs(localtime-sunrise) / DaySize
elif((abs(localtime-sunset) < DaySize)):
    CorrectionFactor=abs(localtime-sunset) / DaySize
elif((abs(localtime-sunrise) > DaySize)or(abs(sunset - localtime) > DaySize)):
    CorrectionFactor=1
print(CorrectionFactor)
```
